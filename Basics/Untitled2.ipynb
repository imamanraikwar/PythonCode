```python
from sys import stdin




def spiralPrint(mat, nRows, mCols):
    #Your code goes here
    temp= 0
    rs = 0
    re = nRows
    cs = 0
    ce = mCols
    r=0
    n= nRows * mCols
    #print(n)
    while temp < n :
        #print(temp)
        for i in range(rs,rs+1):
            for j in range(cs,ce):
                print('I')
                print(mat[i][j],end=' ')
                temp+=1
                #print(temp,'first')
                if temp == n:
                    break
            if temp == n:
                break
        
        rs+=1
        for j in range(ce-1,ce):
            for i in range(rs,re):
                print('II')
                print(mat[i][j],end=' ')
                temp+=1
                #print(temp,'second')
                if temp == n:
                    break
            if temp == n:
                break
        
        ce -= 1
        for i in range(re-1,re):
            for j in range(ce-1,cs-1,-1):
                print('III')
                print(mat[i][j],end=' ')
                temp+=1
                print(temp,'3rd')
                if temp == n:
                    break
            if temp == n:
                break
        
        re-=1
        
        for j in range(cs,cs+1):
            for i in range(re-1,rs-1,-1):
                print(mat[i][j],end=' ')
                temp+=1
        cs+=1
        































#Taking Input Using Fast I/O
def take2DInput() :
    li = [int(x) for x in input().split()]
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(input())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    spiralPrint(mat, nRows, mCols)
    print()

    t -= 1
```
